The same movie can have multiple genres

```python
movies['genres'][0]
```

Define a function to grab only one of the genres (the first)

```python
def grab_top_genre(row):
    try:
        a_json = json.loads(row)
        df_json = json_normalize(a_json)
        top_genre = df_json.iloc[0,1]
    except:
        top_genre = 'Unknown'
    return top_genre
```

```python
movies['top_genre']=movies['genres'].apply(grab_top_genre)
```
